# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,10,20)
y = x * np.random.randn(20) + 2

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x,y, 'k--');
ax.plot(x+1.2,y+.9, 'or')
ax.set(
    title = 'simple plot',
    xlabel = 'x-axis',
    ylabel = 'y-axis'
);
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(14,10))
axes[0,0].plot(x,y, 'k--')
axes[0,0].set(
    title = 'plot',
    ylabel = 'y-axis')

axes[0,1].scatter(x,y, c='r')
axes[0,1].set(
    title = 'scatter')

axes[1,0].bar(x,y)
axes[1,0].set(
    title = 'bar',
    xlabel = 'x-axis',
    ylabel = 'y-axis')

axes[1,1].hist(np.random.randn(40))
axes[1,1].set(
    title = 'histrogram',
    xlabel = 'x-axis');
```

```python
car_data = pd.read_csv('data\car-sales.csv')
car_data
```

```python
car_data['Price'] = car_data['Price'].str.replace('\D', '', regex=True).str[:-2].astype(int)
car_data
```

```python
car_data['Sale Date'] = pd.date_range('1/1/2022', periods=len(car_data))
car_data
```

```python
car_data['Total Sales'] = car_data['Price'].cumsum()
car_data
```

```python
car_data.plot(x='Sale Date', y='Total Sales');
```

```python
car_data.plot(x='Odometer (KM)', y='Price', kind='scatter');
```

```python
heart_disease = pd.read_csv('data\heart-disease.csv')
heart_disease.head()
```

```python
heart_disease['age'].plot.hist(bins=30);
```

```python
over_50 = heart_disease[heart_disease['age'] >= 50].reset_index(drop=True)
over_50.head()
```

### Full customize

```python
fig, axes = plt.subplots(figsize=(10,12), nrows=2, sharex=True)
scatter = axes[0].scatter(
        x=over_50['age'],
        y=over_50['chol'],
        c=over_50['target'],
        cmap='winter',
        s=50,
)
axes[0].set(
    title = 'Heart Disease and Cholesterol Levels',
    ylabel = 'Cholesterol'
)
axes[0].legend(*scatter.legend_elements(), title='Target',fontsize=12)
axes[0].axhline(over_50['chol'].mean(), linestyle = '--')
axes[0].set_ylim([100,450]) 

scatter = axes[1].scatter(
        x=over_50['age'],
        y=over_50['thalach'],
        c=over_50['target'],
        cmap='summer',
        s=50,
)
axes[1].set(
    title = 'Heart Disease and Max Heart Rate',
    xlabel = 'Age',
    ylabel = 'Max Heart Rate'
)
axes[1].legend(*scatter.legend_elements(), title='Target',fontsize=12)
axes[1].axhline(over_50['thalach'].mean(), linestyle = '--')
 
axes[1].set_xlim([48,75]) # use the same x-axis


fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
fig.savefig('heart-analysis-plot.png')
```
